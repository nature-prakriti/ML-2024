```python
def find_argmin(arr):
  index = 0
  max = arr[0]
  for i in range(len(arr)):
    if (arr[i] < max):
      max = arr[i]
      index = i
  return index

arr = np.array([3, 1, 7, 2, 8, 4, 10])
argmin_index = find_argmin(arr)
print(f"Index of the minimum element using custom_argmin: {argmin_index}")

# Verify the result using np.argmin function
numpy_argmin_index = np.argmin(arr)
print(f"Index of the minimum element using np.argmin: {numpy_argmin_index}")
```

```python
# Answer 11
import pandas as pd
import numpy as np

data = {
    'a': [1, 1, 1, 1],
    'b': [1, 4, 6, 8],
    'c': [1, 2, 5, 6],
    'd': [1, 5, 2, 7],
    'e': [1, 2, 1, 5]
}

df = pd.DataFrame(data)

df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)  # axis=1, sum across cols

df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

count = 0
col_LT8 = []
col_GT8 = []
for i in range(len(df)):
  if (df['g'][i] == 'LT8'):
    col_LT8.append(df['f'][i])
    count += 1
  else:
    col_GT8.append(df['f'][i])

print(f"Updated DataFrame: {df}")
print("Number of rows where 'g' is 'LT8':", count)
print("Standard deviation of 'f' where 'g' is 'LT8':", np.std(col_LT8))
print("Standard deviation of 'f' where 'g' is 'GT8':", np.std(col_GT8))
```

```python
# Answer 8

class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        self.marks = {}  # Dictionary to store marks for subjects

    def add_marks(self, subject, marks):
        self.marks[subject] = marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

# Create student objects and add marks
student1 = Student("Student 1", 1)
student1.add_marks("Maths", 100)
student1.add_marks("English", 70)

student2 = Student("Student 2", 2)
student2.add_marks("Maths", 85)
student2.add_marks("English", 75)

# Create branch objects and add students
branch1 = Branch("Branch 1")
branch1.add_student(student1)
branch1.add_student(student2)

branch2 = Branch("Branch 2")

# Create year objects and add branches
year_2022 = Year(2022)
year_2022.add_branch(branch1)
year_2022.add_branch(branch2)

# Create the database object and add years
student_database = Database()
student_database.add_year(year_2022)

# Accessing student information
print("Year:", student_database.years[0].year)
for branch in student_database.years[0].branches:
    print("Branch:", branch.name)
    for student in branch.students:
        print(f"Roll Number: {student.roll_number}, Name: {student.name}, Marks: {student.marks}")
```

```python
# Answer 6
import sympy as sp

x, y = sp.symbols('x y')

f = x**2 * y + y**3 * sp.sin(x)

gradient_f = [sp.diff(f, var) for var in (x, y)]

print("Gradient of f(x, y):", gradient_f)
```

```python
# Answer 5
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

gradient_f = jax.grad(f, argnums=(0, 1))

x_val = 2.0
y_val = 3.0

grad_result = gradient_f(x_val, y_val)

print("Gradient evaluated using JAX:", grad_result)
```
